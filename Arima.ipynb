#### Preliminary analysis of data and stationarity / ACF / PCF plots for AS14.01

```python
df = pd.read_csv('data/data_arima.csv')
df = df[df['mood'].notna()]
```

```python
p_values = [0,1,2,3,4,5,6]
d_values = [0,1,2]
q_values = [0,1,2]

warnings.filterwarnings("ignore")
```

```python
""" skip this cell! instead load data in next cell """

df_results = pd.DataFrame(columns=['p', 'd', 'q', 'mse', 'r2', 'corr', 'tr_na', 'tr', 'tst'], index=df.id.unique()) # init

# Do p,d,q experiment for each person, for each dataset
for person in df.id.unique():
    print(person)
    # Get mood set
    mood = getPersonMoodArimaSet(df, person)
    
    # Results for all p,d,q orders
    pdq_results = do_experiment_ARIMA_find_pdq_values(mood, p_values, d_values, q_values, doPrint=False)

    pdq_results.sort_values('mse', inplace = True)
    pdq_results = pdq_results.reset_index()

    # Save best order
    df_results.loc[person] = pdq_results.loc[0]
    
df_results.to_csv('results/arima.csv')
```

```python
df_results = pd.read_csv('results/arima.csv',index_col=0)
df_results.index.rename('id', inplace=True)

# Plot arima fitted values and forecasting for each person
for person in df.id.unique():
    print(person)
    # Get mood set
    mood = getPersonMoodArimaSet(df, person)

    # Get arima order
    pdq = (df_results.loc[person].p, df_results.loc[person].d, df_results.loc[person].q)
    
    path = 'arima/' + person
    perform_ARIMA(mood, pdq, figureName=path)
```
