```python
print(f'Training set has {len(train)} data points')
print(f'Test set has {len(test)} data points\n')

print(f'Training set has {len(train.srch_id.unique())} searches')
print(f'Test set has {len(test.srch_id.unique())} searches\n')

amt_train_random = len(train.srch_id.where(train["random_bool"]==1).unique())
amt_train_ordered = len(train.srch_id.where(train["random_bool"]==0).unique())
amt_test_random = len(test.srch_id.where(test["random_bool"]==1).unique())
amt_test_ordered = len(test.srch_id.where(test["random_bool"]==0).unique())
print(f'Training set has {amt_train_random} random searches ({(amt_train_random/len(train))}%) and {amt_train_ordered} ordered searches({(amt_train_ordered/len(train))}%)')
print(f'Test set has {amt_test_random} random searches ({(amt_test_random/len(test))}%) and {amt_test_ordered} ordered searches({(amt_test_ordered/len(test))}%)')
```
